# Model building and evaluation 

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor# machine learning algorithm
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split # model evaluationn
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score# Accuracy  
from sklearn.preprocessing import LabelEncoder
```

```python
dfm=pd.read_csv(r"C:\Users\91964\Documents\codesoft\project_2\hh2_IMDb_Movies_India.csv")
dfm.head()
```

```python
dfm["Director"].unique()
```

```python
le=LabelEncoder()
dfm["Genre"]=le.fit_transform(dfm["Genre"])
for label,i in enumerate(le.classes_):
    print(f"{label}:{i}")
```

> <b> label guide for Genre

```python
dfm["Director"]=le.fit_transform(dfm["Director"])
for label,i in enumerate(le.classes_):
    print(f"{label}:{i}")
```

> label guide for director

```python
fet=["Duration","Genre","Votes","Director"]
x=dfm[fet]
y=dfm["Rating"]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
xtrain
```

```python
ytrain
```

# Linear Regression Model

```python
lm=LinearRegression()
lm.fit(xtrain,ytrain)
ypred=lm.predict(xtest)
```

```python
print("MAE",mean_absolute_error(ytest,ypred))
print("MSE",mean_squared_error(ytest,ypred))
print("r2 Score",r2_score(ytest,ypred))
```

```python
lm.predict(xtest.iloc[[1]])
```

```python
ytest.iloc[1]
```

> <b> The regression model shows weak predictive ability, with an MAE of 0.59 and an MSE of 0.94, indicating that predictions deviate from actual values by around half a unit on average with some larger errors. The very low R¬≤ score (0.030) highlights that the model explains less than 3% of the variance in the target variable, meaning it is only slightly better than using the mean as a prediction

## RandomForest Regressor 

```python
rm=RandomForestRegressor(n_estimators=100,random_state=42)
rm.fit(xtrain,ytrain)
ypredr=rm.predict(xtest)
```

```python
print("MAE",mean_absolute_error(ytest,ypredr))
print("MSE",mean_squared_error(ytest,ypredr))
print("r2 Score",r2_score(ytest,ypredr))
```

```python
rm.predict(xtest.iloc[[1]])
```

```python
ytest.iloc[1]
```

> <b> ‚ÄúThe Random Forest Regressor achieved a Mean Absolute Error (MAE) of 0.51 and a Mean Squared Error (MSE) of 0.84, indicating that the model‚Äôs predictions deviate from the actual values by about half a unit on average. However, the R¬≤ score of 0.13 suggests that the model is able to explain only 13% of the variance in the target variable. This indicates that while the Random Forest model reduces prediction errors to some extent, it does not provide a strong fit for the data, and further improvements may be needed through feature engineering, parameter tuning, or trying alternative algorithms.‚Äù

## SVR

```python
svr=SVR(kernel='rbf',degree=3,gamma='scale', C=100, epsilon=0.1)
svr.fit(xtrain,ytrain)
ypreds=svr.predict(xtest)
```

```python
print("MAE",mean_absolute_error(ytest,ypreds))
print("MSE",mean_squared_error(ytest,ypreds))
print("r2 Score",r2_score(ytest,ypreds))
```

```python
svr.predict(xtest.iloc[[1]])
```

```python
ytest.iloc[1]
```

> <b> ‚ÄúThe Support Vector Regressor (SVR) produced a Mean Absolute Error (MAE) of 0.58 and a Mean Squared Error (MSE) of 0.98, meaning that the predictions deviate from the actual values by more than half a unit on average. However, the R¬≤ score is ‚àí0.003, which is lower than zero, indicating that the SVR model performs worse than a simple baseline predictor (such as predicting the mean of the target). This shows that the SVR model does not capture the underlying data patterns effectively and is not suitable in its current form without further tuning or feature improvements

# Decision TreeRegressor

```python
dr=DecisionTreeRegressor(criterion='squared_error',
    max_depth=5,
    min_samples_split=10,
    min_samples_leaf=4,
    max_features='sqrt',
    random_state=42)
dr.fit(xtrain,ytrain)
ypredd=dr.predict(xtest)
```

```python
print("MAE",mean_absolute_error(ytest,ypredd))
print("MSE",mean_squared_error(ytest,ypredd))
print("r2 Score",r2_score(ytest,ypredd))
```

```python
dr.predict(xtest.iloc[[1]])
```

```python
ytest.iloc[1]
```

> <b> Based on the evaluation metrics, the Decision Tree Regressor shows a Mean Absolute Error (MAE) of 0.5666 and a Mean Squared Error (MSE) of 0.8636, which indicates that the model‚Äôs predictions deviate moderately from the actual values. However, the R¬≤ score of 0.1119 reveals that the model is able to explain only about 11% of the variance in the target variable. This suggests that while the model is producing predictions with some level of accuracy, its overall explanatory power is quite low. In other words, the Decision Tree is not effectively capturing the underlying patterns in the dataset, and improvements such as hyperparameter tuning or using more advanced ensemble techniques may be necessary to achieve better performance.

# Conclusion 

## Among the models tested, linear regression provided very weak predictive power with only a small portion of variance explained. Support Vector Regression performed the worst, with almost no ability to capture meaningful patterns. A single decision tree showed slight improvement but still remained limited due to its tendency to overfit and poor generalization. In contrast, ensemble methods like random forests are generally the most effective choice, as they combine multiple trees to capture nonlinear relationships and usually deliver stronger and more reliable predictive performance compared to the other approaches.
